# Préparation des données

```python
# Descriptif des variables 

# Age
print("\n info variable Age \n")
print(df["Age"].describe())

# Sex
print("\n info variable Sex \n")
print(df["Sex"].value_counts())

# Job
print("\n info variable Job \n")
print(df["Job"].value_counts())

# Housing
print("\n info variable Housing \n")
print(df["Housing"].value_counts())

# Saving accounts
print("\n info variable Saving accounts \n")
print(df["Saving accounts"].value_counts())

# Checking accounts
print("\n info variable Checking account \n")
print(df["Checking account"].value_counts())

# Credit amount
print("\n info variable Credit amount \n")
print(df["Credit amount"].describe())

# Duration
print("\n info variable Duration \n")
print(df["Duration"].describe())

# Purpose
print("\n info variable Purpose \n")
print(df["Purpose"].value_counts())
```

```python
# Travail sur les variables

# Variable Sex
df.Sex.replace({'male': 0, 'female': 1},inplace=True)

# Variable Job 
df["Job"]=df["Job"].astype("category")

# Variable Housing
df=df.join(pd.get_dummies(df["Housing"],prefix="cat_"))
df = df.drop("Housing", axis=1)

# Variable Saving accounts
df = df.drop("Saving accounts", axis=1)

# Checking account 
df = df.drop("Checking account", axis=1)

# Purpose
df = df.drop("Purpose", axis=1)

# Variable Risk
df.Risk.replace({'good': 0, 'bad': 1},inplace=True)
```

```python
# vérification des changements
df.info()
```

# Modélisation

## Préparation de la base

```python
# Importations des modules néccésaires 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import classification_report
import xgboost as xgb

# Définition de la target
Y=df["Risk"]
X=df.drop("Risk", axis=1)

# Split train test
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size= 0.2 , random_state=123)
```

```python
X.to_csv(r"C:\Users\yanis\01 Projets\01 Python Projects\Projet_Tech_Sprint\02_Données\0_Données_entrainement\X.csv")
X_train.to_csv(r"C:\Users\yanis\01 Projets\01 Python Projects\Projet_Tech_Sprint\02_Données\0_Données_entrainement\X_train.csv")
X_test.to_csv(r"C:\Users\yanis\01 Projets\01 Python Projects\Projet_Tech_Sprint\02_Données\0_Données_entrainement\X_test.csv")
Y.to_csv(r"C:\Users\yanis\01 Projets\01 Python Projects\Projet_Tech_Sprint\02_Données\0_Données_entrainement\Y.csv")
y_train.to_csv(r"C:\Users\yanis\01 Projets\01 Python Projects\Projet_Tech_Sprint\02_Données\0_Données_entrainement\y_train.csv")
y_test.to_csv(r"C:\Users\yanis\01 Projets\01 Python Projects\Projet_Tech_Sprint\02_Données\0_Données_entrainement\y_test.csv")
```

## Regression logistique

```python
clf=LogisticRegression(C=1.0)
clf.fit(X_train,y_train)
y_pred_logistique=clf.predict(X_test)

print(classification_report(y_test, y_pred_logistique))
```

## Random Forest

```python
clf_rf=RandomForestClassifier(n_jobs=-1,random_state=321)
clf_rf.fit(X_train,y_train)
y_pred_rf=clf_rf.predict(X_test)

print(classification_report(y_test, y_pred_rf))
```

## Gradient boosting

```python
clf_gb=GradientBoostingClassifier()
clf_gb.fit(X_train,y_train)
y_pred_gb=clf_gb.predict(X_test)

print(classification_report(y_test, y_pred_gb))
```

## XGboost

```python
# Repréparation des données
X_xgb, X_valid, y_xgb, y_valid = train_test_split(X, Y, test_size= 0.1 , random_state=123)
X_train, X_test, y_train, y_test = train_test_split(X_xgb, y_xgb, test_size= 0.2 , random_state=123)

train = xgb.DMatrix(data = X_train, label= y_train,enable_categorical=True)
test =  xgb.DMatrix(data = X_test, label = y_test,enable_categorical=True)
valid = xgb.DMatrix(data = X_valid, label = y_valid,enable_categorical=True)
```

```python
params = {'booster':'gbtree', 'learning_rate': 1, 'objective': 'binary:logistic'}

xgb1 = xgb.train(params = params,
                dtrain = train, 
                num_boost_round= 1000,
                evals= [(train, 'train'), (test, 'eval')])

preds = xgb1.predict(test)

xgbpreds= pd.Series(np.where(preds > 0.5, 1, 0))
```

```python
print(classification_report(y_test, xgbpreds))
```
